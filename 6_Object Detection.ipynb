## two-stage detector
#### 첫 번째 단계: 영역제안
- 이미지 내에서 객체가 존재할 가능할 가능성이 있는 영역, 후보 영역을 추출
- 후보 영역은 수백 수천개가 될 수도 있음

#### 두 번째 단계: 분류와 바운딩 박스 회귀
- 첫 번째 단계에서 추출된 후보 영역 각각을 딥러닝 모델에 입력으로 제공하여 해당 영역 내의 객체를 분류
- 또한 정확한 객체의 위치를 나타내기 위해 경계 상자의 사이즈와 위치를 조절하는 작업도 함께 수행

### R-CNN(Region with Convolutional Neural Networks)

1. 영역제안
- 이전의 객체탐지는 이미지 전체를 슬라이딩 윈도우 방식으로 탐색하여 객체를 찾음
- 이 방식은 연산량이 많아져 효율적이지 못함
- 이를 해결하기 위해 영역 제안 알고리즘, 즉 이미지 내에서 객체가 있을 법한 영역을 빠르게 식별하는 방법이 도입
- R-CNN에서 영역 제안은 선택적 검색 방식으로 진행이 되었으며 영역 제안 방식의 주요 절차는 다음과 같음
```mermaid
초과분할 --> 계층적 그룹화 --> 전략의 다양화 --> 필터링 제안
```
2. 계층적 그룹화
- 세그먼트들을 특정 기준에 따라 병합하는 과정
  - 병합 전략: 색상, 질감, 사이즈 및 형태 유사성을 기반으로 가장 유사한 세그먼트들을 우선적으로 병합

3. 전략의 다양화
- 세그먼트를 병합할 때 단일 전략만을 사용하는 것이 아닌, 여러 가지 전략을 동시에 사용하여 다양한 스케일과 모양의 영역을 포착할 수 있도록 함

4. 필터링 제안
- 많은 수의 영역 제안 중에서 가장 유용한 제안만을 선택
  - 사이즈 기반 필터링: 너무 작거나 큰 영역 제안 제외
  - 다양성 최대화: 서로 비슷한 영역 제안 중에서는 하나만 선택
  - 제안 수 제한: R-CNN에서는 약 2천개의 영역 제안만을 선택하여 사용

#### CNN을 활용한 특징 추출
- R-CNN은 선택적 검색으로 추출된 영역 제안들을 합성곱 신경망을 통해 분석
- 합성곱 층은 이미지의 지역적인 특징을 학습하고, 완전 연결 층은 이미지의 전역적인 정보를 학습
- 이러한 층들의 조합을 통해, R-CNN은 이미지 내의 복잡한 패턴과 특성을 효과적으로 파악하며, 이 정보는 후속 과정에서 객체의 분류와 경계 상자의 정밀화를 위해 사용

#### SVM을 활용한 분류
- R-CNN에서는 마지막 단계로 서포트 벡터 머신(Support Vector Machine, SVM) 분류기를 사용하여 각각의 영역 제안을 특정 객체 클래스로 분류

#### NMS, IOU
- 바운딩 박스의 중복된 검출을 처리하기 위하여 추가적인 단계가 필요하여, 이때 사용되는 것이 NMS
- 작동 방식
  1. 모든 바운딩 박스를 객체 점수(신뢰도)에 따라 내림차순 정렬
  2. 사용자가 설정한 임계치 이상의 점수를 가진 박스 선택
  3. 다른 경계 상자 제거
- 이때 중복되는 정도를 나타내는 점수는 IoU값을 기준으로 하며 0.5가 넘어가면 겹치는 박스로 설정
$$IoU = \frac{\text{교집합의 영역}}{\text{합집합의 영역}}$$
