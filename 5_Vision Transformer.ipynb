### RNN의 단점(바닐라RNN)
1. GPU 연산의 장점인 병렬화가 어려움
2. tanh 활성화 함수를 쓰기 때문에 기울기 소실/폭발이 생김 -> 멀리 있는 정보와의 관계 계산이 어려워짐 (장기 의존성 문제)

### LSTM(Long Short Term Memory)
- 한번씩 Hidden Layer를 지날때 마다 가중치(w)에 따라 필요없는 state와 필요한 state를 구분하여 다음 Node로 넘긴다
- Hidden State 안에 셀 상태를 추가
    - 입력 게이트
    - 삭제 게이트
    - 출력 게이트

### GRU(Gated Recurrent Unit)
- LSTM의 Hidden State의 연산량을 줄임
    - 업데이트 게이트
    - 리셋 게이트

### Word Embedding
- 단어의 Vector화
- Ex) I = (1.0, 3.4, 2.7), love = (2.4, 1.7, 4.5)

#### Attention
- 각각의 값의 중요도를 이용한 "무언가"를 사용
- input 값을 더했을때 1이 나오게 각각의 input을 설정하고 그 값들을 기반으로 output을 예측하기 위해 가중치를 설정함

여기서 RNN의 단점 1번이 등장함<br>
Ex) 난 학교 에서 공부해<br>
라는 문장을 생성할 때, "에서"라는 단어가 나오기 위해서는 "학교"라는 단어를 먼저 추론해야하는데,<br>
"학교"라는 단어가 나와야 "에서"라는 단어를 얻을 수 있으므로 GPU 병렬 처리로 결과를 얻을 수 없음<br>
