```python
import cv2

image = cv2.imread('like_lenna.png', cv2.IMREAD_GRAYSCALE)

if image is not None:
    print('Load Success')
else:
    print('Load Fail')
print(f"변수 타입: {type(image)}")
```

```python
from google.colab.patches import cv2_imshow # imshow 옵션을 따로 불러와서 사용한다

cv2_imshow(image)

print(f"이미지 배열의 형태: {image.shape}")
```

```python
image_small = cv2.resize(image, (100, 100)) # 크기 조정
cv2_imshow(image_small)
```

```python
image_fliped = cv2.flip(image, 0) # 상하 반전
cv2_imshow(image_fliped)
```

```python
image_fliped = cv2.flip(image, 1) # 좌우 반전
cv2_imshow(image_fliped)
```

```python
height, width = image.shape
matrix = cv2.getRotationMatrix2D((width/2, height/2), 90, 1)
image_rotated = cv2.warpAffine(image, matrix, (width, height))
cv2_imshow(image_rotated)
```

```python
matrix = cv2.getRotationMatrix2D((width/2, height/2), 30, 1)
image_rotated = cv2.warpAffine(image, matrix, (width, height), borderValue=200)
cv2_imshow(image_rotated)
```

```python
import numpy as np
space = np.zeros((500,1000), dtype = np.uint8)
line_color = 255
space = cv2.line(space, (100,100), (800,300), line_color, 3, 1)

cv2_imshow(space)
```

```python
space = np.zeros((500, 1000), dtype=np.uint8)
color = 255
space = cv2.circle(space, (600,200), 100, color, 4, 1)

cv2_imshow(space)
```

```python
space = np.zeros((768, 1388), dtype= np.uint8)
line_color = 255
space = cv2.rectangle(space, (500, 200), (800, 400), line_color, 5, 1)

cv2_imshow(space)
```

```python
space = np.zeros((768, 1388), dtype= np.uint8)
line_color = 255
space = cv2.ellipse(space, (500, 300), (300, 200), 0, 90, 250, line_color, 4)

cv2_imshow(space)
```

```python
space = np.zeros((768,1388), dtype=np.uint8)
color = 255
obj1 = np.array([[300, 500], [500, 500], [400, 600], [200, 600]])
obj2 = np.array([[600, 500], [800, 500], [700,200]])
space = cv2.polylines(space, [obj1], True, color, 3)
space = cv2.fillPoly(space, [obj2], color, 1)

cv2_imshow(space)
```

### cv2.imwrite( ) 함수
- 행렬을 영상파일로 저장
- 확장자에 따라서 JPG, BMP, PNG, TIF, PPM 등의 영상파일 포맷 저장 가중

#### cv2.IMWRITE_JPEQ_QUALITY
- param: 0 ~ 100
- JPG 파일 화질, 높은 값일수록 화질 좋음

#### cv2.IMWRITE_PNG_COMPRESSION
- param: 0 ~ 9
- PNG 파일 압축 레벨, 높은 값일수록 용량은 적어지고, 압축 시간이 길어짐

#### cv2.IMWRITE_PXM_BINARY
- param: 0 or 1
- PPM, PGM 파일의 이진 포맷 설정

```python
import cv2

title1, title2 = '16bit unchanged', '32bit unchanged'
color2unchanged1 = cv2.imread("/content/drive/MyDrive/2025_1_Colab/ComputerVision_Pattern/images/room.png", cv2.IMREAD_UNCHANGED)
color2unchanged2 = cv2.imread("/content/drive/MyDrive/2025_1_Colab/ComputerVision_Pattern/images/cafe.jpeg", cv2.IMREAD_UNCHANGED)

if color2unchanged1 is not None and color2unchanged2 is None: # 예외처리
    raise Exception("Error")

print("16/32비트 영상 행렬 좌표(10, 10) 화소값")
print(title1, '원소 자료형', type(color2unchanged1[10, 10, 0])) # 원소 자료형
print(title1, '화소값(3원소)', color2unchanged1[10, 10])        # 한 화소 값 표시
print(title2, '원소 자료형', type(color2unchanged1[10, 10, 0]))
print(title2, '화소값(3원소)', color2unchanged1[10, 10])
print()

cv2_imshow(color2unchanged1) # 16비트 정수형 행렬 영상 표시
cv2_imshow(color2unchanged2 * 255).astype('uint8') # 정수형 변홯 ㅜ
cv2.waitKey(0)
```
