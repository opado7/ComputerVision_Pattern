```python
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras

from tensorflow.keras import layers
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D,MaxPooling2D,Flatten,Dropout,Dense
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras.optimizers import Adam
```

```python
# CIFAR-10 데이터 세트 로드
(train_images, train_labels), (test_images, test_labels) = cifar10.load_data()
print(train_images.shape, train_labels.shape)
print(test_images.shape, test_labels.shape)
```

```python
train_images = train_images / 255.0
test_images = test_images / 255.0
```

```python
val_images = train_images[45000:]
val_labels = train_labels[45000:]

train_images = train_images[:45000]
train_labels = train_labels[:45000]
```

```python
cnn = Sequential()
cnn.add(Conv2D(96, (11, 11), strides = (4, 4), activation='sigmoid', padding = 'same', input_shape=(32, 32, 3)))
cnn.add(MaxPooling2D((2, 2), strides=(2, 2)))
cnn.add(Conv2D(256, (5, 5), padding = 'same', activation='sigmoid'))
cnn.add(MaxPooling2D((2, 2), strides=(2, 2)))
cnn.add(Conv2D(384, (3, 3), padding = 'same', activation='sigmoid'))
cnn.add(Conv2D(384, (3, 3), padding = 'same', activation='sigmoid'))
cnn.add(Conv2D(256, (3, 3), padding = 'same', activation='sigmoid'))
cnn.add(MaxPooling2D((2, 2), strides=(2, 2)))
cnn.add(Flatten())
cnn.add(Dense(4096, activation='sigmoid'))
cnn.add(Dense(4096, activation='sigmoid'))
cnn.add(Dense(1000, activation='sigmoid'))
cnn.add(Dense(10, activation='softmax'))
```

```python
cnn.compile(optimizer=Adam(learning_rate=0.001), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
cnn.fit(train_images, train_labels, epochs=5, validation_data=(val_images, val_labels))
```

```python
res = cnn.evaluate(test_images, test_labels)
print('정확도 = {:.3f}%'.format(res[1]*100))
```
